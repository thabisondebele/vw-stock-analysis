Data Quality Note — Volume Zeros
243 trading days recorded zero volume. For a major exchange-listed stock like Volkswagen, true zero-volume days are implausible. These are treated as missing data and replaced using forward fill (previous trading day's volume). This affects approximately 3% of records and is documented here for transparency.

```python
print(zero_vol.index.min())   # earliest zero-volume date
print(zero_vol.index.max())   # latest zero-volume date
print(zero_vol.index.year.value_counts().sort_index())  # how many per year
```

Data Quality Decision — Start Date
The 1995–1999 period contains severe volume data gaps (109 zero-volume days in 1995 alone, representing ~44% of that year's trading days). This indicates poor data digitisation from the early period rather than genuine market activity. Analysis has been restricted to 2000 onwards where data quality is reliable. This decision is conservative and prioritises analytical integrity over dataset size.

```python
df = df[df.index.year >= 2000]

print(f'Dataset now runs from {df.index.min().date()} to {df.index.max().date()}')
print(f'Total trading days: {len(df)}')
```

```python
import numpy as np

print('Average volume before fix:', df['Volume'].mean().round(0))

df['Volume'] = df['Volume'].replace(0, np.nan)
df['Volume'] = df['Volume'].ffill()

print('Zero volume days after fix:', (df['Volume'] == 0).sum())
print('Average volume after fix:', df['Volume'].mean().round(0))
```

```python
print('Final missing values:')
print(df.isnull().sum())
print('Date range:', df.index.min(), 'to', df.index.max())
print('Total trading days:', len(df))
```

## Created Volatility_30d

This column was missing from the dataset. It calculates the standard deviation 
of daily returns over a rolling 30-day window, measuring how wildly the stock 
price is swinging at any given point in time.

High value = large unpredictable moves = high risk period (e.g. 2008, COVID-19)  
Low value  = calm steady movement = stable market conditions

30 days because it is responsive enough to catch real events quickly, 
but long enough to smooth out single-day noise.
